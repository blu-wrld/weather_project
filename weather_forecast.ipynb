```python
core_data["temp_max_target"] = core_data.shift(-1)["temp_max"]  ## Create a column target that is depicting the next days temp_max value
core_data["temp_min_target"] = core_data.shift(-1)["temp_min"]  ## Create a column target that is depicting the next days temp_min value
core_data
```

```python
core_data = core_data.iloc[:-1,:].copy() ## Cut off the last row because target is null
core_data
```

```python
## Import machine learning model, using ridge regression, a type of regression that minimizes overfitting
from sklearn.linear_model import Ridge 
reg_max = Ridge(alpha=.1) 
reg_min = Ridge(alpha=.1) 
```

```python
predictors = ["percip", "snow_depth", "temp_max", "temp_min"] ## specify variables used to predic outcome
```

```python
## Divide data set into training and test set, training set is from 1901 to 2021, test set is 2021-2023
train = core_data.loc[:"2021-12-31"] 
test = core_data.loc["2022-01-01":]
```

```python
 ## Fit this model to our data to get target column using training dataset
reg_max.fit(train[predictors], train["temp_max_target"]) 
reg_min.fit(train[predictors], train["temp_min_target"]) 
```

```python
##Use our fitted model to generate predictions on our "test" dataset using predictors
predictions_max = reg_max.predict(test[predictors]) 
```

```python
##Use oue fitted model to generate predictions on our "test" dataset using predictors
predictions_min = reg_min.predict(test[predictors]) 
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
## Absolute of the difference between predicted and actual temp max value. 
mean_absolute_error(test["temp_max_target"], predictions_max) 
```

```python
## Absolute of the difference between predicted and actual temp min value. 
mean_absolute_error(test["temp_min_target"], predictions_min) 
```

```python
## Concatenate actual and predicted temp max values
temp_max_combined = pd.concat([test["temp_max_target"], pd.Series(predictions_max, index = test.index)], axis = 1) #predicted target values are in numpy array need to convert to panda series
temp_max_combined.columns = ["actual", "predictions"] ## Rename the columns
temp_max_combined
```

```python
## Concatenate actual and predicted temp min values
temp_min_combined = pd.concat([test["temp_min_target"], pd.Series(predictions_min, index = test.index)], axis = 1) 
temp_min_combined.columns = ["actual", "predictions"] ## Rename the columns
temp_min_combined
```

```python
## Turn all the steps into a function (temp_max)
def create_temp_max_predicitions(predictors, core_data, reg_max):
    train = core_data.loc[:"2021-12-31"] 
    test = core_data.loc["2022-01-01":] 
    reg_max.fit(train[predictors].astype(float), train["temp_max_target"].astype(float))
    predictions = reg_max.predict(test[predictors])
    error = mean_absolute_error(test["temp_max_target"], predictions)
    temp_max_combined = pd.concat([test["temp_max_target"], pd.Series(predictions, index = test.index)], axis = 1) 
    temp_max_combined.columns = ["actual", "predictions"]
    return error, temp_max_combined 
```

```python
## Turn all the steps into a function (temp_min)
def create_temp_min_predicitions(predictors, core_data, reg_min):
    train = core_data.loc[:"2021-12-31"] 
    test = core_data.loc["2022-01-01":] 
    reg_min.fit(train[predictors], train["temp_min_target"])
    predictions = reg_min.predict(test[predictors])
    error = mean_absolute_error(test["temp_min_target"], predictions)
    temp_min_combined = pd.concat([test["temp_min_target"], pd.Series(predictions, index = test.index)], axis = 1) 
    temp_min_combined.columns = ["actual", "predictions"]
    return error, temp_min_combined 
```

```python
predictors_for_temp_max = ["percip", "snow_depth", "temp_max","temp_min", "month_max", "month/day_max"]
predictors_for_temp_min = ["percip", "snow_depth", "temp_max", "temp_min", "month_min"]
```
